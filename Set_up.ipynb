```python
#Treatment of Categorical Data
#Substituting Values for Categorical Data
for i in range(len(main.loc[:,'BUILDING TYPE'])):
    if(main.loc[i,'BUILDING TYPE']!=main.loc[i,'BUILDING TYPE']):
        main.loc[i,'BUILDING TYPE'] = main.loc[i-1,'BUILDING TYPE']
        
for i in range(len(main.loc[:,'BUILDING TYPE'])):
    if(main.loc[i,'BUILDING_SUBTYPE']!=main.loc[i,'BUILDING_SUBTYPE']):
        main.loc[i,'BUILDING_SUBTYPE'] = main.loc[i-1,'BUILDING_SUBTYPE']
```
